### **[DICAS] - 11 gráficos de PANDAS úteis com uma linha de código**
---

Vamos conhecer a sintaxe para realizar o **plot** de gráficos com o `PANDAS` utilizando apenas _uma_ linha de código.

```python
# importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import warnings

# configurando o ambiente para exibição
%matplotlib inline
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12,7)
warnings.filterwarnings('ignore')
```

```python
# importando de uma "url" o conjunto de dados
!wget https://raw.githubusercontent.com/LearnPythonWithRune/pandas_charts/main/air_quality.csv
```

```python
# visualizando o diretório principal de trabalho
%ls
```

```python
# inspecinando o conjunto de dados
!head air_quality.csv
```

```python
# instanciando o conjunto de dados no objeto "data" e, definindo os seguintes parâmetros:
# index_col=0 e parse_dates=True | aqui estamos definindo que a coluna "datetime" será o nosso índice
data = pd.read_csv('air_quality.csv', index_col=0, parse_dates=True)
```

```python
# visualizando as 5 primeiras observações
data.head()
```

```python
# informações gerais
data.info()
```

#### **Gráficos com PANDAS**

**1. Gráfico de linha simples**

```python
# data.plot(figsize=(12,7));
data.plot();
```

**2. Gráfico de linhas simples isolado**

```python
# data['station_london'].plot(figsize=(12,7));
data['station_london'].plot();
```

```python
# data['station_antwerp'].plot(figsize=(12,7));
data['station_antwerp'].plot();
```

```python
# data.plot(subplots=True, figsize=(12,7));
data.plot(subplots=True);
```

**3. Gráfico de dispersão**

```python
# data.plot.scatter(x='station_london', y='station_paris', alpha=0.25, figsize=(12,7));
data.plot.scatter(x='station_london', y='station_paris', alpha=0.25);
```

**4. Gráfico de caixa**

```python
# data.plot.box(figsize=(12,7));
data.plot.box();
```

**5. Gráfico de área**

```python
data.plot.area(subplots=True);
```

**6. Gráfico de barras**

```python
data.iloc[:15].plot.bar();
```

```python
data.iloc[:15].plot.barh();
```

**7/8. Gráfico de histograma**

```python
data['station_london'].plot.hist();
```

```python
data['station_london'].plot.hist(bins=30);
```

```python
data.plot.hist(alpha=0.5);
```

```python
data.plot.hist(alpha=0.5, bins=30);
```

**9. Gráfico de torta (pizza)**

```python
data.mean().plot.pie();
```

```python
data.count().plot.pie();
```

**10. Gráfico _matriz de dispersão_**

```python
from pandas.plotting import scatter_matrix
```

```python
scatter_matrix(data, alpha=0.2);
```

**11. Eixo y secundário**

```python
data['station_london'].plot();
data['station_london'].cumsum().plot(secondary_y=True);
```

```python
data['station_london'].plot();
data['station_paris'].cumsum().plot(secondary_y=True);
```
